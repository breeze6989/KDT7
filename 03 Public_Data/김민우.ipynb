```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
```

```python

file = [
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\광주 기온.csv", "광주"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\대구 기온.csv", "대구"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\대전 기온.csv", "대전"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\부산 기온.csv", "부산"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\서울 기온.csv", "서울"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\인제 기온.csv", "인제"),
    ("C:\\Users\\KDP-50\\OneDrive\\바탕 화면\\KDT7\\04_PUBLIC_DATA\\team_project\\제주 기온.csv", "제주")
]


fig, axes = plt.subplots(2, 4, figsize=(24, 12))
axes = axes.flatten()[:len(file)]  

for i, (file_path, city_name) in enumerate(file):
    
    data = pd.read_csv(file_path, encoding="EUC-KR")    
    
    data.iloc[:, 1:].plot(kind='bar', width=0.7, color=['steelblue', 'gold', 'crimson', 'slategray'], ax=axes[i])    
    
    axes[i].set_title(f'{city_name} 평균 기온 (1991~2020년)', fontsize=12)
    axes[i].set_ylabel('기온 (°C)', fontsize=10)
    axes[i].set_xlabel('월', fontsize=10)
    axes[i].set_xticks(range(0, 12))
    axes[i].set_xticklabels(['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'], rotation=0)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)
    axes[i].set_ylim(-10, 30)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.suptitle('한국 주요 도시 월별 평균 기온 비교 (1991~2020년)', fontsize=16)
plt.show()
```

```python
data=pd.read_csv('전세계 지표 바다 기온변화(이동평균).csv',encoding="EUC_KR")

plt.figure(figsize=(14, 7))

plt.plot(data['Year'], data['Land_Annual'], color='dimgray', marker='o', label='평균 기온 (°C)')
plt.plot(data['Year'], data['Lowess(5)'], color='crimson', linewidth=2, label='5년 이동 평균')

plt.title('전세계 평균 기온', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('기온 (°C)', fontsize=12)
plt.xticks(range(1880, 2021, 20))
plt.yticks([-1.0, -0.5, 0.0, 0.5, 1.0, 1.5])
plt.legend(fontsize=12)
plt.grid(linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
data=pd.read_csv('우리나라기온변화.csv',encoding="EUC_KR")

plt.figure(figsize=(14, 7))
plt.plot(data['연도'], data['45개 지점'], color='firebrick', marker='o')
plt.title('대한민국 평균 기온', fontsize=20, pad=20)
plt.xlabel('연도', fontsize=12)
plt.ylabel('기온 (°C)', fontsize=12)
plt.xticks(range(1973, 2024, 5))
plt.yticks([-1.0, -0.5, 0.0, 0.5, 1.0, 1.5])
plt.grid(linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
data=pd.read_csv('지표온도.csv',encoding="EUC_KR")

plt.figure(figsize=(14, 8))


for column in data.columns[1:]:
    plt.plot(data['연도'], data[column], label=column)


plt.title('국내 지표온도 연평균', fontsize=20, pad=20)
plt.xlabel('연도', fontsize=12)
plt.ylabel('지표온도(°C)', fontsize=12)
plt.xticks(range(1973, 2024, 5))


plt.legend(fontsize=10, ncol=4, loc='upper center', bbox_to_anchor=(0.5, -0.15))


plt.grid(linestyle='--', alpha=0.5)


plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
data=pd.read_csv('토양수분.csv', encoding="EUC_KR")
plt.figure(figsize=(14, 8))


plt.bar(data['월'] - 0.15, data['2000년대'], width=0.3, label='2000년대', color='paleturquoise')
plt.bar(data['월'] + 0.15, data['2010년대'], width=0.3, label='2010년대', color='dodgerblue')


plt.title('2000년대 vs 2010년대 월별 토양수분 비교', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('토양수분 (%)', fontsize=12)
plt.xticks(data['월'], labels=[f'{month}월' for month in data['월']])


plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```
