```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# lr_def.py 내부 함수 임포트
from lr_def import (
    plot_features_and_target,
    split_train_test,
    regression_kfold_analysis,
    LinearRegression_kfold_analysis,
    plot_regression_results,
    predict_and_evaluate_model,
    # 필요시 전처리 함수도 임포트 가능
    power_transform_dataframe,
    fill_outliers_with_median_modified_zscore,
    MinMaxScaler_dataframe,
    RobustScaler_dataframe,
    StandardScaler_dataframe
)
from sklearn.preprocessing import StandardScaler

#########################################
# [A] 파일 경로 지정
m_ncaa_file   = "../data2/MNCAATourneyDetailedResults.csv"
m_regular_file= "../data2/MRegularSeasonDetailedResults.csv"

w_ncaa_file   = "../data2/WNCAATourneyDetailedResults.csv"
w_regular_file= "../data2/WRegularSeasonDetailedResults.csv"

# 고정 피처 14개 
FIXED_FEATURE_COLS = [
    'WScore','WFGM','WFGA','WFGM3','WFGA3','WFTM','WFTA',
    'WOR','WDR','WAst','WTO','WStl','WBlk','WPF'
]
TARGET_COL = 'now'

#########################################
# [B] Season>2023 / Season=2023 구분 로드 함수
def load_data_by_season(file_path, season_col='Season', wloc_col='WLoc', 
                        season_condition='greater'):
    """
    file_path     : CSV 파일 경로
    season_col    : 시즌 컬럼명('Season')
    wloc_col      : 승리팀 위치 컬럼명('WLoc')
    season_condition : 'greater' → Season>2023, 'equal' → Season=2023
    
    반환: 필터링 & WLoc 제거된 DataFrame
    """
    df = pd.read_csv(file_path)
    
    if season_condition == 'greater':
        df = df[df[season_col] > 2023]
    elif season_condition == 'equal':
        df = df[df[season_col] == 2023]
    else:
        # 필요시 다른 조건 처리
        pass
    
    # WLoc 컬럼 제거
    if wloc_col in df.columns:
        df.drop(columns=[wloc_col], inplace=True)
    return df

#########################################
# [C] 결합 후, WTeamID 기준 그룹화 -> 숫자형 합산 + now(등장 횟수) + 정렬
def combine_and_aggregate(df1, df2, group_col='WTeamID'):
    """
    두 DataFrame(df1, df2)를 세로 concat 후,
    group_col 기준으로 그룹화:
      - 숫자형 합산
      - 그룹 크기를 now 컬럼에
      - now 기준 내림차순 정렬
    """
    combined = pd.concat([df1, df2], ignore_index=True)

    size_df = combined.groupby(group_col).size().reset_index(name='now')
    sum_df = combined.groupby(group_col).sum(numeric_only=True).reset_index()

    final_df = pd.merge(sum_df, size_df, on=group_col, how='left')

    # now 내림차순 정렬
    final_df.sort_values(by='now', ascending=False, inplace=True)
    final_df.reset_index(drop=True, inplace=True)

    return final_df

#########################################
# [D] 모델 학습 (Season>2023 데이터)용 함수
def do_regression_analysis(df, feature_cols, target_col='now', label='M'):
    """
    df           : 학습용 DataFrame (Season>2023)
    feature_cols : 회귀에 사용할 컬럼들(14개 고정)
    target_col   : 'now'
    label        : M / W
    반환: (학습 완료된 4가지 모델) - (ridge, lasso, elastic, linear)
          그리고 (적절하다면) 스케일러
    """
    # feature_cols 중 실제 존재하는 컬럼만
    valid_features = [col for col in feature_cols if col in df.columns]
    if len(valid_features) == 0:
        print(f"[{label}] 유효한 피처가 없어 학습 불가.")
        return None, None, None, None, None

    # 결측 제거
    use_cols = valid_features + [target_col]
    df = df.dropna(subset=use_cols).copy()
    if len(df) < 5:
        print(f"[{label}] 데이터가 너무 적어서 회귀분석 불가.")
        return None, None, None, None, None

    X = df[valid_features]
    y = df[target_col]

    # (선택) 스케일링
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # train/test split
    x_train, x_test, y_train, y_test = split_train_test(
        pd.DataFrame(X_scaled, columns=valid_features),
        y, test_size=0.2, random_state=42
    )

    # 1) Ridge, Lasso, ElasticNet
    ridge_df, lasso_df, elastic_df, ridge_model, lasso_model, elastic_model = regression_kfold_analysis(
        x_train, x_test, y_train, y_test,
        alpha_list=[0.01, 0.1, 0.5, 1, 5, 10]
    )

    print(f"\n--- {label} Ridge plot ---")
    plot_regression_results(ridge_df)
    print(f"--- {label} Lasso plot ---")
    plot_regression_results(lasso_df)
    print(f"--- {label} ElasticNet plot ---")
    plot_regression_results(elastic_df)

    # 2) LinearRegression
    linear_df, linear_model = LinearRegression_kfold_analysis(x_train, x_test, y_train, y_test)
    print(f"--- {label} LinearRegression plot ---")
    plot_regression_results(linear_df)

    # 3) 예측 성능
    print(f"\n==== 최종 예측 성능(Ridge) : [{label}] ====")
    predict_and_evaluate_model(ridge_model, x_test, y_test)

    print(f"==== 최종 예측 성능(Lasso) : [{label}] ====")
    predict_and_evaluate_model(lasso_model, x_test, y_test)

    print(f"==== 최종 예측 성능(ElasticNet) : [{label}] ====")
    predict_and_evaluate_model(elastic_model, x_test, y_test)

    print(f"==== 최종 예측 성능(Linear) : [{label}] ====")
    predict_and_evaluate_model(linear_model, x_test, y_test)

    # 모델 4개 & 스케일러 반환
    return (ridge_model, lasso_model, elastic_model, linear_model, scaler)


#########################################
# [E] 2023 데이터에 대해 예측값 산출
def predict_on_2023_data(df_2023, feature_cols, target_col='now', label='M',
                        trained_models=None, scaler=None):
    """
    df_2023       : Season=2023 집계가 끝난 DataFrame
    feature_cols  : 학습 때 사용한 컬럼 (14개)
    target_col    : 'now'
    label         : M / W
    trained_models: (ridge, lasso, elastic, linear)
    scaler        : 학습 시 fit한 StandardScaler

    - df_2023에 동일한 전처리(결측 제거 + scaler.transform) 후
      4개 모델 각각 predict
    - 결과 DataFrame 형태로 출력
    """
    if trained_models is None or scaler is None:
        print(f"[{label}] 학습 모델 또는 스케일러가 없습니다. 예측 불가.")
        return

    # 사용할 피처가 실제 df_2023에 있는지 확인
    valid_features = [col for col in feature_cols if col in df_2023.columns]
    if len(valid_features) == 0:
        print(f"[{label}] 2023 데이터에 유효한 피처 없음.")
        return

    # 결측 제거
    use_cols = valid_features + [target_col]
    df_2023 = df_2023.dropna(subset=use_cols).copy()
    if df_2023.empty:
        print(f"[{label}] 2023 데이터가 비어있어 예측 불가.")
        return

    X_2023 = df_2023[valid_features]
    y_2023 = df_2023[target_col]  # 실제 now

    # 스케일 transform
    X_scaled_2023 = scaler.transform(X_2023)

    # 모델 4개 분리
    ridge_model, lasso_model, elastic_model, linear_model = trained_models

    # 예측
    pred_ridge   = ridge_model.predict(X_scaled_2023)
    pred_lasso   = lasso_model.predict(X_scaled_2023)
    pred_elastic = elastic_model.predict(X_scaled_2023)
    pred_linear  = linear_model.predict(X_scaled_2023)

    result_df = pd.DataFrame({
        'WTeamID': df_2023['WTeamID'],  # 식별용
        'Actual_now': y_2023,
        'Ridge_pred'  : pred_ridge,
        'Lasso_pred'  : pred_lasso,
        'Elastic_pred': pred_elastic,
        'Linear_pred' : pred_linear
    })

    print(f"\n[{label}] Season=2023 예측 결과:")
    print(result_df.head(20))  # 상위 20개 정도 표시 (필요시 조정)

    # 필요하다면 RMSE, R2 등도 간단히 계산 가능
    # 여기서는 예시로 R2, RMSE 한번씩 계산
    from sklearn.metrics import r2_score, root_mean_squared_error
    def rmse(y_true, y_pred):
        return root_mean_squared_error(y_true, y_pred, )

    print(f"\n[{label}] --- Ridge 2023 ---")
    print("R2:", r2_score(y_2023, pred_ridge), "/ RMSE:", rmse(y_2023, pred_ridge))

    print(f"\n[{label}] --- Lasso 2023 ---")
    print("R2:", r2_score(y_2023, pred_lasso), "/ RMSE:", rmse(y_2023, pred_lasso))

    print(f"\n[{label}] --- ElasticNet 2023 ---")
    print("R2:", r2_score(y_2023, pred_elastic), "/ RMSE:", rmse(y_2023, pred_elastic))

    print(f"\n[{label}] --- Linear 2023 ---")
    print("R2:", r2_score(y_2023, pred_linear), "/ RMSE:", rmse(y_2023, pred_linear))


#########################################
# [F] 실행 흐름
def main():

    # 1) 남자(M) Season>2023 로드 & 결합 & 학습
    m_ncaa_gt   = load_data_by_season(m_ncaa_file,   season_condition='greater')
    m_regular_gt= load_data_by_season(m_regular_file,season_condition='greater')
    # 결합 / group
    m_df_agg_gt = combine_and_aggregate(m_ncaa_gt, m_regular_gt, group_col='WTeamID')

    # 남자(M) 학습
    m_models = do_regression_analysis(m_df_agg_gt, FIXED_FEATURE_COLS, target_col=TARGET_COL, label='M')
    # m_models = (ridge, lasso, elastic, linear, scaler)

    # 2) 남자(M) Season=2023 로드 & 결합 & 예측
    m_ncaa_2023    = load_data_by_season(m_ncaa_file,   season_condition='equal')
    m_regular_2023 = load_data_by_season(m_regular_file,season_condition='equal')
    m_df_agg_2023  = combine_and_aggregate(m_ncaa_2023, m_regular_2023, group_col='WTeamID')

    if m_models and None not in m_models:
        ridge_m, lasso_m, elastic_m, linear_m, m_scaler = m_models
        # 예측
        predict_on_2023_data(
            df_2023 = m_df_agg_2023,
            feature_cols = FIXED_FEATURE_COLS,
            target_col   = TARGET_COL,
            label        = 'M',
            trained_models = (ridge_m, lasso_m, elastic_m, linear_m),
            scaler        = m_scaler
        )
    else:
        print("[M] 학습 모델이 없어서 2023 데이터 예측 불가.\n")

    # 3) 여자(W) Season>2023 로드 & 결합 & 학습
    w_ncaa_gt   = load_data_by_season(w_ncaa_file,   season_condition='greater')
    w_regular_gt= load_data_by_season(w_regular_file,season_condition='greater')
    w_df_agg_gt = combine_and_aggregate(w_ncaa_gt, w_regular_gt, group_col='WTeamID')

    w_models = do_regression_analysis(w_df_agg_gt, FIXED_FEATURE_COLS, target_col=TARGET_COL, label='W')

    # 4) 여자(W) Season=2023 로드 & 결합 & 예측
    w_ncaa_2023    = load_data_by_season(w_ncaa_file,   season_condition='equal')
    w_regular_2023 = load_data_by_season(w_regular_file,season_condition='equal')
    w_df_agg_2023  = combine_and_aggregate(w_ncaa_2023, w_regular_2023, group_col='WTeamID')

    if w_models and None not in w_models:
        ridge_w, lasso_w, elastic_w, linear_w, w_scaler = w_models
        # 예측
        predict_on_2023_data(
            df_2023 = w_df_agg_2023,
            feature_cols = FIXED_FEATURE_COLS,
            target_col   = TARGET_COL,
            label        = 'W',
            trained_models = (ridge_w, lasso_w, elastic_w, linear_w),
            scaler        = w_scaler
        )
    else:
        print("[W] 학습 모델이 없어서 2023 데이터 예측 불가.\n")

    print("\n=== 모든 작업이 완료되었습니다. ===")

#########################################
# 실제 실행
if __name__ == "__main__":
    main()

```
